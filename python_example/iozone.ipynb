```python
import re

initial_writers = 'Children see throughput for  1 initial writers'
readers = 'Children see throughput for  1 readers'

with open("D:/pull log/Linux_Storage.txt") as f:
    content = f.readlines()
    


def check_regex_commands(content,regex_pattern):
    expect_log=list()
    for log in content:
        try:
            expect_log += re.findall(regex_pattern, log)
        except:
            pass
    return expect_log          

#content)

# find patten
Children_initial_writers = 'Children see throughput for  1 initial writers\D+\d+.\d+'
Children_readers   = 'Children see throughput for  1 readers  \D+\d+.\d+'
Children_random_writers  = 'Children see throughput for 1 random writers\D+\d+.\d+'
Children_random_readers  = 'Children see throughput for 1 random readers\D+\d+.\d+'



ciw = check_regex_commands(content,Children_initial_writers)
cr = check_regex_commands(content,Children_readers)
crw= check_regex_commands(content,Children_random_writers)
crr= check_regex_commands(content,Children_random_readers)


def element_extraction(patten):
    words=[]
    for i in range(0,len(patten)):
        word = float(patten[i].split('=')[1].strip())
        words.append(word) 
    
    words.append(round(min(words),2))
    words.append(round(max(words),2))
    words.append(round((sum(words)/5),2))
    return words

e1 = element_extraction(ciw)
e2 = element_extraction(cr)
e3 = element_extraction(crw)
e4 = element_extraction(crr)



print(e1)
print(e2)
print(e3)
print(e4)
```

```python
def element(test):
    string_list = ['1','2','3','4','5','min','max','ave']
    
    dictionary = dict(zip(string_list,test))
    return dictionary

# Children_initial_writers 
print('Children_initial_writers\n' , element(e1))
# Children_readers 
print('Children_readers\n', element(e2))
# Children_random_writers
print('Children_random_writers\n', element(e3))
#Children_random_readers
print('Children_random_readers\n' ,element(e4))
```
